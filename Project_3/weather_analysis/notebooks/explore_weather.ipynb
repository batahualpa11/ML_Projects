```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2], ignore_index=True)

weather.to_csv("../data/weather/weather_data.csv", index=False)
```

```python
# TODO: view first 5 rows of your data
weather.head()
```

```python
# TODO: describe your numeric columns
weather.describe()
```

```python
# TODO: display the null values in your dataframe
weather.isnull().sum()
```

```python
# TODO: BEGIN YOUR EDA!

# Set the style for all plots
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 8))

# Create a line plot for average temperature over years
plt.subplot(2, 2, 1)
plt.plot(weather['year'], weather['temp_avg'], marker='o', linestyle='-', color='red')
plt.title('Average Temperature Over Years')
plt.xlabel('Year')
plt.ylabel('Temperature (C)')
plt.grid(True)

# Create a line plot for minimum humidity over years
plt.subplot(2, 2, 2)
plt.plot(weather['year'], weather['hum_min'], marker='o', linestyle='-', color='blue')
plt.title('Minimum Humidity Over Years')
plt.xlabel('Year')
plt.ylabel('Humidity (%)')
plt.grid(True)

# Create a line plot for average wind speed over years
plt.subplot(2, 2, 3)
plt.plot(weather['year'], weather['wind_avg'], marker='o', linestyle='-', color='green')
plt.title('Average Wind Speed Over Years')
plt.xlabel('Year')
plt.ylabel('Wind Speed (m/s)')
plt.grid(True)

# Create a bar plot for maximum rainfall by year
plt.subplot(2, 2, 4)
plt.bar(weather['year'], weather['rain_max'], color='purple')
plt.title('Maximum Rainfall by Year')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

# Create a scatter plot for max rain vs max humidity
plt.figure(figsize=(10, 6))
plt.scatter(weather['hum_max'], weather['rain_max'], color='orange', alpha=0.7)
plt.title('Relationship Between Max Humidity and Max Rainfall')
plt.xlabel('Maximum Humidity (%)')
plt.ylabel('Maximum Rainfall (mm)')
plt.grid(True)

# Add a trend line
z = np.polyfit(weather['hum_max'], weather['rain_max'], 1)
p = np.poly1d(z)
plt.plot(weather['hum_max'], p(weather['hum_max']), "r--", alpha=0.8)

# Calculate and display the correlation coefficient
correlation = weather['hum_max'].corr(weather['rain_max'])
plt.annotate(f'Correlation: {correlation:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.show()

# Create a scatter plot for average wind speed vs average temperature
plt.figure(figsize=(10, 6))
plt.scatter(weather['temp_avg'], weather['wind_avg'], color='teal', alpha=0.7)
plt.title('Relationship Between Average Temperature and Average Wind Speed')
plt.xlabel('Average Temperature (C)')
plt.ylabel('Average Wind Speed (m/s)')
plt.grid(True)

# Add a trend line
z = np.polyfit(weather['temp_avg'], weather['wind_avg'], 1)
p = np.poly1d(z)
plt.plot(weather['temp_avg'], p(weather['temp_avg']), "r--", alpha=0.8)

# Calculate and display the correlation coefficient
correlation = weather['temp_avg'].corr(weather['wind_avg'])
plt.annotate(f'Correlation: {correlation:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.show()
```
