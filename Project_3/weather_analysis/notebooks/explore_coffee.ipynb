```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull().sum()
```

```python
# TODO: drop your null values
coffee = coffee.dropna()
```

```python
# TODO: Select only Minas Gerais

minas_gerais = coffee[coffee['subdivision'] == 'Minas Gerais']
```

```python
# TODO: BEGIN YOUR EDA!
# BEGIN YOUR EDA!

# Set the style for all plots
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(15, 10))

# 1. Plot bearing trees over time
plt.subplot(2, 2, 1)
plt.plot(minas_gerais['year'], minas_gerais['bearing_trees'], marker='o')
plt.title('Bearing Trees in Minas Gerais Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Bearing Trees')
plt.grid(True)

# 2. Plot bearing hectares over time
plt.subplot(2, 2, 2)
plt.plot(minas_gerais['year'], minas_gerais['bearing_hectares'], marker='o', color='green')
plt.title('Bearing Hectares in Minas Gerais Over Time')
plt.xlabel('Year')
plt.ylabel('Bearing Hectares')
plt.grid(True)

# 3. Plot 60kgs bags over time
plt.subplot(2, 2, 3)
plt.plot(minas_gerais['year'], minas_gerais['60kgs_bag'], marker='o', color='red')
plt.title('60kg Bags in Minas Gerais Over Time')
plt.xlabel('Year')
plt.ylabel('Million 60kg Bags')
plt.grid(True)

# 4. Scatter plot of bearing trees vs 60kgs bags
plt.subplot(2, 2, 4)
plt.scatter(minas_gerais['bearing_trees'], minas_gerais['60kgs_bag'])
plt.title('Relationship: Bearing Trees vs 60kg Bags')
plt.xlabel('Number of Bearing Trees')
plt.ylabel('Million 60kg Bags')
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate correlation coefficients
print("\nCorrelation between bearing_trees and 60kgs_bag:")
print(minas_gerais['bearing_trees'].corr(minas_gerais['60kgs_bag']))

print("\nCorrelation between bearing_hectares and 60kgs_bag:")
print(minas_gerais['bearing_hectares'].corr(minas_gerais['60kgs_bag']))

# Distribution plots
plt.figure(figsize=(15, 5))

# Distribution of bearing trees
plt.subplot(1, 3, 1)
sns.histplot(minas_gerais['bearing_trees'], kde=True)
plt.title('Distribution of Bearing Trees')
plt.xlabel('Number of Bearing Trees')

# Distribution of bearing hectares
plt.subplot(1, 3, 2)
sns.histplot(minas_gerais['bearing_hectares'], kde=True, color='green')
plt.title('Distribution of Bearing Hectares')
plt.xlabel('Bearing Hectares')

# Distribution of 60kgs bags
plt.subplot(1, 3, 3)
sns.histplot(minas_gerais['60kgs_bag'], kde=True, color='red')
plt.title('Distribution of 60kg Bags')
plt.xlabel('Million 60kg Bags')

plt.tight_layout()
plt.show()
```
